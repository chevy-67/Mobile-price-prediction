```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error,r2_score,root_mean_squared_error,mean_absolute_error,accuracy_score
from sklearn.ensemble import RandomForestRegressor
import joblib
```

#**Dataset** **Preprocessing**

```python
df = pd.read_csv("https://drive.google.com/uc?export=downlaod&id=1BafUUknI8AbzySY907wyw-f2wNMdZbk3")
print(df.columns)
```

```python
df.Touchscreen.unique()
df = df[df.Touchscreen == 'Yes']
df.drop(columns=['Touchscreen'],inplace=True)
print(df.columns)
```

```python
print(df['Wi-Fi'].unique())
print(df.Bluetooth.unique())
print(df.GPS.unique())
print(df['3G'].unique())
```

```python
df.drop(columns=['Wi-Fi','Bluetooth','GPS','3G'],inplace=True)
```

```python
df.drop(columns=['Unnamed: 0','Name'],inplace = True)
df.columns
```

```python
#df.rename(columns={'Battery capacity (mAh)':'battery_capacity_mAh','Screen size (inches)':'screen_size_inch','Resolution x':'Resolution_x','Resolution y':'Resolution_y',})
df.columns = df.columns.str.replace(' ','_').str.lower()
df.head()
```

```python
df.rename(columns={'4g/_lte':'4G/LTE','operating_system':'OS','ram_(mb)':'RAM_(MB)','internal_storage_(gb)':'internal_storage_(GB)'},inplace = True)
df['4G/LTE'] = df['4G/LTE'].apply(lambda x: 1 if x=='Yes' else 0)
df.head()
```

```python
df['4G/LTE'].unique()
```

```python
df.drop(columns='4G/LTE',inplace=True)
df.head()
```

```python
print(df.brand)
print(df.model)
```

```python
df.to_csv("cleaned_dataset_mobile_prediction.csv",index=False)
```

```python
le = LabelEncoder()
df['OS'] = le.fit_transform(df['OS'])
#df['brand'] = le.fit_transform(df['brand'])
df.head()
```

```python
df.brand.unique()
```

```python
df.head()
```

#**Model** **Building**

```python
x = df.drop(columns=['price','model','brand'])
y = df.price
print(x.head())
print(y.head())
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)

lr = LinearRegression()
lr_model = lr.fit(x_train,y_train)
```

```python
rf_model = RandomForestRegressor(n_estimators=100,random_state=42)
rf_model.fit(x_train,y_train)
```

#**Model Evalution**

###**Using Linear Regression**










```python
y_pred = lr_model.predict(x_test)

mse = mean_squared_error(y_test,y_pred)
rmse = root_mean_squared_error(y_test,y_pred)
r2score = r2_score(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)


print("Accuracy using Linear Regression")
print("Mean Squared Error\t: ",mse)
print("Root Mean Squared Error\t: ",rmse)
print("R2 Score\t\t: ",r2score)
print("Mean Absolute Error\t: ",mae)
```

###**Using Random Forest Regressor**

```python
y_pred_rf = rf_model.predict(x_test)

mse_rf = mean_squared_error(y_test,y_pred_rf)
rmse_rf = root_mean_squared_error(y_test,y_pred_rf)
r2score_rf = r2_score(y_test,y_pred_rf)
mae_rf = mean_absolute_error(y_test,y_pred_rf)

print("Accuracy using Random Forest Regressor")
print("Mean Absolute Error\t: ",mse_rf)
print("Root Mean Squared Error\t: ",rmse_rf)
print("R2 Score\t\t: ",r2score_rf)
print("Mean Absolute Error\t: ",mae_rf)
print("Accuracy : ",accuracy_score(y_test,y_pred_rf,))
```

```python
joblib.dump(lr_model,'mobile_linearRegression_model.joblib')
joblib.dump(rf_model,'mobile_randomForest_model.joblib')
```
